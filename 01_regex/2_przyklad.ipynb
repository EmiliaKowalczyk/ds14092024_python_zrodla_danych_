#### Dopasowanie zachłanne (*ang. greedy matching*)

Dopasowanie zachłanne dotyczy powtórzeń (czyli metaznaków `*`, `+`, `?`, '{}').

Wyrażenie `wn[0-9]*` oznacza ciąg 'wn', po którym bezpośrednio występuje 0 lub więcej cyfr. Do wzorca `wn[0-9]*` pasuje więc nie tylko fragment tekstu 'wn123', ale również 'wn', 'wn1' oraz 'wn12'. Pomimo to dla ciągu 'Bob Brown: bobbro`wn123`@home.co.uk' otrzymaliśmy wyłącznie jedno dopasowanie - 'wn123'. Takie zachowanie nazywamy **dopasowaniem zachłannym**. Dopasowanie zachłanne oznacza, że w przypadku niejednoznaczności dopasowania, dopasowywana jest zawsze maksymalna możliwa liczba powtórzeń. Jest to zachowanie domyślne dla wszystkich metaznaków opisujących liczbę powtórzeń (czyli `*`, `+`, `?`, `{}`). Możne je zmienić za pomocą odpowiedniej flagi.

#### Dopasowanie leniwe (ang. *lazy matching*)

Zmianę domyślnego zachowania z zachłannego na leniwe można zrealizować za pomocą:
- flagi (`**U**ngreedy)
- metaznaku `?`

**Wyrażenie regularne:** `wn[0-9]*?` dopasuje najmniejszą możliwą liczbę powtórzeń (tutaj pomiędzy 0 a dowolną liczbą powtrzeń `[0-9]` czyli 0)

**Oczekiwany wynik:**

**Wyrażenie regularne:** `wn[0-9]??` dopasuje najmniejszą możliwą liczbę powtórzeń [0-9] (tutaj pomiędzy 0 a 1 powtórzenie `[0-9]` czyli 0)



**Wyrażenie regularne:** `wn[2-4]?` dopasuje najmniejszą możliwą liczbę powtórzeń [0-9] (tutaj pomiędzy dwa a cztery powtórzenia `[0-9]` czyli 2)

#### Tryb wielolinijkowy (*ang. multiline*) i flagi

Tryb wielolnijkowy dotyczy metaznaków `^`, `$`.

Domyślny tryb to tryb wielolinijkowy. W trybie wielolinikowym metaznak `^` oznacza początek nowej linii, a metaznak `$` oznacza koniec nowej linii. To zachowanie można zmienić za pomocą odpowiedniej flagi. 



Po zmianie zachowania metaznak `^` oznaczać będzie początek całego tekstu, a metaznak `$` oznaczać będzie koniec całego tekstu.

Zestawienie flag:

| Flaga             | Opis                                                                                          | Przykład/Zastosowanie                                           |
|-------------------|-----------------------------------------------------------------------------------------------|-----------------------------------------------------------------|
| `g (`**g**`lobal)`      | Wyszukuje wszystkie wystąpienia wzorca w tekście, a nie tylko pierwsze dopasowanie.           | `/pattern/g`                                                   |
| `m (`**m**`ultiline)`   | Umożliwia dopasowanie początku i końca każdej linii w wieloliniowym tekście.                  | `/^start/m`                                                    |
| `i (`**i**`nsensitive)` | Ignoruje wielkość liter przy dopasowaniu.                                                     | `/pattern/i`                                                   |
| `x (e`**x**`tended)`    | Pozwala na bardziej przejrzyste wzorce przez ignorowanie białych znaków i komentarzy.         | `/pattern/x`                                                   |
| `s (`**s**`ingle line)`      | Umożliwia dopasowanie znaku nowej linii przez kropkę `.`.                                     | `/pattern/s`                                                   |
| `u (`**u**`nicode)`     | Umożliwia dopasowanie wzorców zgodnie ze standardem Unicode.                                  | `/pattern/u`                                                   |
| `U (`**U**`ngreedy)`    | Odwraca domyślne zachowanie chciwego dopasowania, powodując, że operator `*`, `+` i `{}` stają się niechciwe (dopasowują jak najmniej znaków). | `/pattern/U`                                                   |
| `A (`**A**`nchored)`       | Wymusza dopasowanie wzorca tylko na początku tekstu.                                          | `/pattern/A`                                                   |
| `J (`**J**`changed)`       | Pozwala na wielokrotne użycie we wzorcu tej samej nazwy grupy.                                | `/pattern/J`                                                   |
| `D (`**D**`ollar end only)`| Wymusza, aby znak dolara `$` dopasowywał tylko koniec tekstu, a nie koniec linii w trybie wieloliniowym. | `/pattern/$`                                                   |

#### Grupy przechwytujące (*ang. capturing groups*)

Zwróć uwagę, że poza samym dopasowaniem, oddzielnie dostajemy też dopasowanie samej grupy. Domyślnie grupy w wyrażeniach regularnych są właśnie tzw. grupami przechwytującymi. Oznacza to, że do dopasowanej grupy można się później odwołać. Takie odwołanie pozwala na jej ewentualne usunięcie lub zmodyfikowanie w oryginalnym tekście.

#### Grupy nieprzechwytujące (ang. non-capturing groups)

Jeżeli nie interesuje nas zawartość grupy możemy oznaczyć ją jako nieprzechwytującą. W tym celu na samym początku definicji grupy należy umieścić symbol `?:`. Oznaczmy drugą grupę z poprzedniego przykładu jako grupę nieprzechwytującą.

Wyrażenie regularne: `([a-zA-Z]+):\/\/(?:[a-zA-Z]+)\.([a-zA-Z0-9]+)\.([a-zA-Z]{2,})`



Widzimy, że w dopasowaniach brak jest tym razem grupy www.

#### 19. Metasekwencje ze znakiem `\`

Dialekty posiadają też metasekwencje, którymi nazywamy skrótowe zapisy reprezentujące często pojawiające się wyrażenia regularne. Najczęściej metasekwencje składają się z otwierającego metasekwencje znaku `/` po którym następuje jakaś litera. W ten sposób:
- metasekwencja `\d` zastępuję wyrażenie `[0-9]`
- metasekwencja `\w` (tzw. znak słowny) zastępuje wyrażenie `[a-zA-Z0-9_]`

Zestawienie najpopularniejszych metasekwencji:

| Metasekwencja | Opis                          | Regex                       |
|----------------|-------------------------------|-----------------------------|
| `\d`           | Dowolna cyfra                 | `[0-9]`                     |
| `\D`           | Dowolny znak, który nie jest cyfrą | `[^0-9]`                 |
| `\w`           | Dowolny znak alfanumeryczny lub podkreślnik | `[A-Za-z0-9_]` |
| `\W`           | Dowolny znak niealfanumeryczny | `[^A-Za-z0-9_]`            |
| `\s`           | Dowolny znak biały (spacja, tabulator, nowa linia) | `[ \t\r\n\f]` |
| `\S`           | Dowolny znak, który nie jest białym znakiem | `[^ \t\r\n\f]` |

#### 21. Przeszukiwanie wsteczne (*ang. lookbehind*)
Jest to temat zaawansowany, ale tylko jego brakuje nam do kompletu, dlatego tutaj króciutko o nim wspominamy.

Przeszukiwanie wsteczne (lookbehind) jest to typ grupy, która pozwala nam odwoływać się do fragmentu występującego przed dopasowaniem. Jest to ustandaryzowny sposób robienia tego co wcześniej zrobilismy już za pomocą znaku resetowania `\K`. Grupę przeszukiwania wstecznego oznacza się `(?<=)` podając po znaku `=` wzorzec, który ma wystąpić przed właściwym dopasowaniem. Przypomnijmy sobie przykład ze znakiem resetu `\K`.



Interesuje nas wyłącznie fraza `ma` występujące po frazie `Ala `. Regex ma postać: `Ala \Kma`. Regex realizujący identyczne dopasowanie za pomocą grupy przeszukiwania wstecznego (*lookbehind*) będzie miał postać `(?<=Ala )ma`:



#### 23. Przeszukiwanie przyszłe (*ang. lookahead*)

Grupa przeszukiwania przyszłego (*lookahead*) działa w identyczny sposób jak grupa przeszukiwania wstecznego (lookbehind), tylko że odwołuje się do tekstu następującego po właściwym dopasowaniu. Grupę dopasowania przyszłego oznacza się `(?=)`. Znajdźmy tylko takie frazy `ma`, które następują przed frazą ` Alę`. Regex realizujący takie dopasowanie będzie postaci: `ma(?= Alę)`.



#### 24. Negatywne przeszukiwanie wsteczne/przyszłe (ang. *negative lookbehind, negative lookahead*)

Tak jak wyszukujemy dopasowania w przeszukiwaniu wstecznym i przyszłym tak możemy też wprowadzić do nich negacje. Czyli zamiast dopasowywać, to wykluczać jakiś wzorzec (przed lub za) dopasowaniem. W tym celu zamieniamy znak `=` na `!`, czyli:
- `(?<!)` - negatywne przeszukianie wsteczne
- `(?!)` - negatywne przeszukiwanie przyszłe
